```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cars_engage_2022_updated.csv")
df
```

```python
data = list(df.Fuel_Type)
data = list(set(data)) 
l = []
dataset = list(df.Fuel_Type)
for val in data:
    l.append(dataset.count(val))

li = l
plt.figure("")
plt.barh(data,l,color="blue")
plt.xlabel("No of Cars")
plt.ylabel("Fuel Types")
plt.title("Fuel Types vs No of Cars")

for index, value in enumerate(li):
    plt.text(value, index, str(value))
    
plt.show()
```
